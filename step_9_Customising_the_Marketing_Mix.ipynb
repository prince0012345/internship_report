```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Example: Load or simulate ausActivDesc
# ausActivDesc = pd.read_csv("ausActivDesc.csv")

# Simulated data (for demonstration)
np.random.seed(42)
ausActivDesc = pd.DataFrame({
    'spendpppd': np.random.gamma(shape=2., scale=50., size=200),  # Skewed spending data
})

# Simulate cl12_3 as in previous step
# Let's say half are in Segment 3
ausActivDesc['cl12_3'] = ["Segment 3" if i < 100 else "Not Segment 3" for i in range(200)]

# Convert to category (optional)
ausActivDesc['cl12_3'] = ausActivDesc['cl12_3'].astype("category")

# Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='cl12_3', y='spendpppd', data=ausActivDesc,
            notch=True, width=0.6)

# Set y-axis to log scale
plt.yscale("log")

# Label
plt.ylabel("AUD per person per day")
plt.title("Spending Distribution by Segment")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate a dataset
np.random.seed(42)
n = 200
ausActivDesc = pd.DataFrame({
    'cl12_3': ['Segment 3']*100 + ['Not Segment 3']*100,
    'bookHotel': np.random.choice([0, 1], size=n),
    'bookFlight': np.random.choice([0, 1], size=n),
    'bookCar': np.random.choice([0, 1], size=n)
})

# Convert 'cl12_3' to category type
ausActivDesc['cl12_3'] = ausActivDesc['cl12_3'].astype('category')

# Step 1: Select columns starting with 'book'
book_cols = [col for col in ausActivDesc.columns if col.startswith('book')]

# Step 2: Calculate % of 1s per group for each column
prop_df = ausActivDesc.groupby('cl12_3')[book_cols].mean() * 100
prop_df = prop_df.T  # transpose for better plotting

# Step 3: Plot grouped bar chart
ax = prop_df.plot(kind='barh', xlim=(-2, 102), figsize=(8, 4), legend=True)

# Labels and styling
plt.xlabel("Percent")
plt.title("Booking Behavior by Segment")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulated dataset
np.random.seed(1)
n = 200
ausActivDesc = pd.DataFrame({
    'cl12_3': ['Segment 3'] * 100 + ['Not Segment 3'] * 100,
    'infoInternet': np.random.choice([0, 1], size=n),
    'infoBrochure': np.random.choice([0, 1], size=n),
    'infoFriend': np.random.choice([0, 1], size=n),
    'infoTV': np.random.choice([0, 1], size=n),
})

# Ensure cl12_3 is treated as a category
ausActivDesc['cl12_3'] = ausActivDesc['cl12_3'].astype('category')

# Step 1: Select 'info' columns
info_cols = [col for col in ausActivDesc.columns if col.startswith("info")]

# Step 2: Group by segment and calculate % of 1s
info_props = ausActivDesc.groupby('cl12_3')[info_cols].mean() * 100
info_props = info_props.T  # Transpose for horizontal bar plot

# Step 3: Plot
plt.figure(figsize=(8, 5))
ax = info_props.plot(kind='barh', xlim=(-2, 102), figsize=(8, 5), legend=True)

# Styling
plt.xlabel("Percent")
plt.title("Information Sources by Segment")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Simulated data
np.random.seed(2)
n = 200
ausActivDesc = pd.DataFrame({
    'cl12_3': ['Segment 3'] * 100 + ['Not Segment 3'] * 100,
    'TV.channel': np.random.choice(['BBC', 'CNN', 'FOX', 'Local'], size=n)
})

# Step 1: Create contingency table
contingency = pd.crosstab(ausActivDesc['cl12_3'], ausActivDesc['TV.channel'])

# Step 2: Plot mosaic chart
plt.figure(figsize=(8, 6))
mosaic(contingency.stack(), title='TV Channel by Segment', gap=0.01)
plt.xticks(rotation=90)  # Equivalent to par(las = 2) in R
plt.xlabel("")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
