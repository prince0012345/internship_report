```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better aesthetics
sns.set_style("whitegrid")

# Load the dataset
data = pd.read_csv('EV Stats.csv')

# Clean column names for simplicity
data.columns = [
    'Sl_No', 'State',
    'Two_Wheelers_L1_L2', 'Two_Wheelers_L2_CMVR', 'Two_Wheelers_250W',
    'Three_Wheelers_L5_Slow', 'Three_Wheelers_L5',
    'Passenger_Cars_M1', 'Buses', 'Total'
]

# Select relevant columns (exclude Sl_No and Total)
vehicle_columns = [
    'Two_Wheelers_L1_L2', 'Two_Wheelers_L2_CMVR', 'Two_Wheelers_250W',
    'Three_Wheelers_L5_Slow', 'Three_Wheelers_L5',
    'Passenger_Cars_M1', 'Buses'
]

# Ensure numeric data
for col in vehicle_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce').fillna(0)

# Create a multiline plot
plt.figure(figsize=(10, 7))

# Plot each vehicle category as a separate line
for col in vehicle_columns:
    plt.plot(data['State'], data[col], marker='o', label=col.replace('_', ' '))

# Customize the plot
plt.title('Electric Vehicle Sales by Category Across Indian States', fontsize=14, pad=10)
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Vehicles Sold', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(title='Vehicle Category', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Save the plot
plt.savefig('ev_sales_multiline.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("whitegrid")

# Hypothetical survey data based on Lucknow and Anand studies
survey_data = pd.DataFrame({
    'Age_Group': ['18-25']*350 + ['26-35']*300 + ['36-45']*200 + ['46+']*150,
    'Vehicle_Type': (
        ['Two-Wheelers']*245 + ['Passenger_Cars']*70 + ['Three-Wheelers']*35 +
        ['Two-Wheelers']*180 + ['Passenger_Cars']*90 + ['Three-Wheelers']*30 +
        ['Two-Wheelers']*80 + ['Passenger_Cars']*100 + ['Three-Wheelers']*20 +
        ['Two-Wheelers']*75 + ['Passenger_Cars']*45 + ['Three-Wheelers']*30
    ),
    'Category': (
        ['EV']*210 + ['ICE']*35 + ['EV']*60 + ['ICE']*10 + ['EV']*30 + ['ICE']*5 +
        ['EV']*135 + ['ICE']*45 + ['EV']*68 + ['ICE']*22 + ['EV']*24 + ['ICE']*6 +
        ['EV']*40 + ['ICE']*40 + ['EV']*20 + ['ICE']*80 + ['EV']*12 + ['ICE']*8 +
        ['EV']*23 + ['ICE']*52 + ['EV']*5 + ['ICE']*40 + ['EV']*6 + ['ICE']*24
    )
})

# Load EV Stats.csv for South India data
data = pd.read_csv('EV Stats.csv')
data.columns = [
    'Sl_No', 'State', 'Two_Wheelers_L1_L2', 'Two_Wheelers_L2_CMVR', 'Two_Wheelers_250W',
    'Three_Wheelers_L5_Slow', 'Three_Wheelers_L5', 'Passenger_Cars_M1', 'Buses', 'Total'
]
vehicle_columns = [
    'Two_Wheelers_L1_L2', 'Two_Wheelers_L2_CMVR', 'Two_Wheelers_250W',
    'Three_Wheelers_L5_Slow', 'Three_Wheelers_L5', 'Passenger_Cars_M1', 'Buses'
]
for col in vehicle_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce').fillna(0)

# South India subset
south_states = ['Karnataka', 'Tamil Nadu', 'Kerala', 'Andhra Pradesh', 'Telangana']
south_data = data[data['State'].isin(south_states)][['State'] + vehicle_columns]

# 1. Bar chart of vehicle type preferences by age group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Group', hue='Vehicle_Type', data=survey_data, palette='viridis')
plt.title('Vehicle Type Preferences by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Consumers', fontsize=12)
plt.legend(title='Vehicle Type')
plt.tight_layout()
plt.savefig('age_vehicle_bar.png', bbox_inches='tight')
plt.close()

# 2. Heatmap of EV sales in South India
plt.figure(figsize=(10, 6))
sns.heatmap(south_data.set_index('State')[vehicle_columns], annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('EV Sales by Vehicle Type in South Indian States', fontsize=14)
plt.xlabel('Vehicle Category', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.tight_layout()
plt.savefig('south_india_heatmap.png', bbox_inches='tight')
plt.show()
```
